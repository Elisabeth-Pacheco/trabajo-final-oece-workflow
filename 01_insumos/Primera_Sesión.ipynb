## Idea central

Programar es **explicitar reglas de transformación**. En términos empíricos, transformamos datos. En términos conceptuales, transformamos abstracciones en operaciones. En términos categoriales, componemos funciones sobre objetos bien definidos.

## 1. Programar en ciencias sociales: una primera formulación rigurosa

En ciencias sociales, los datos no son “cosas dadas”, sino registros construidos institucionalmente. Una variable como ingreso, años de educación o participación ciudadana no agota la realidad: apenas codifica una dimensión de una relación social. Por ello, programar bien implica tres cuidados: (i) **cuidado conceptual**: saber qué representa cada variable; (ii) **cuidado lógico**: saber qué transformación estamos aplicando; y (iii) **cuidado técnico**: saber si el código respeta tipos, nombres y estructura.

Desde la teoría de categorías, podemos introducir una intuición útil desde el inicio, pues un **objeto** puede ser una tabla, una lista o una observación; un **morfismo** puede ser una función que transforma ese objeto; y un **pipeline** es una composición ordenada de funciones. Nuestro objetivo no es solo “hacer correr código”, sino construir transformaciones comprensibles, verificables y reproducibles.

## 2. Objetos, variables y tipos

Una variable en Python es un nombre que referencia un objeto almacenado en memoria. Ese objeto tiene un **tipo**, y el tipo importa porque determina qué operaciones son válidas. En investigación social, esto es crucial:

- una edad puede representarse como `int`;
- un ingreso monetario como `float`;
- una categoría territorial como `str`;
- una condición lógica como `bool`.

Una mala tipificación genera errores analíticos antes incluso de que aparezcan errores estadísticos.

```python
# Ejemplos básicos con sentido social

department = "Cusco"          # str
year = 2024                   # int
average_income = 1480.75      # float
is_urban = False              # bool

print("department     =", department, "| tipo:", type(department))
print("year           =", year, "| tipo:", type(year))
print("average_income =", average_income, "| tipo:", type(average_income))
print("is_urban       =", is_urban, "| tipo:", type(is_urban))
```

```python
# Convenciones de nombres: claridad antes que brevedad

household_income = 1650.50
years_of_education = 11
participates_in_assembly = True

print("Ingreso del hogar      :", household_income)
print("Años de educación      :", years_of_education)
print("Participa en asamblea  :", participates_in_assembly)
```

```python
# Conversión de tipos y cuidado analítico

raw_income = "2350.80"
income = float(raw_income)

print("Valor original :", raw_income, "| tipo:", type(raw_income))
print("Valor convertido:", income, "| tipo:", type(income))

# Error típico: querer sumar texto y número
# print(raw_income + 10)  # Esto produciría un TypeError
```

## 3. Estructuras de datos para representar relaciones sociales

La elección de la estructura no es trivial. Cada estructura organiza la información de una forma distinta.

- **lista**: secuencia ordenada y mutable;
- **diccionario**: pares clave-valor;
- **tupla**: secuencia ordenada e inmutable;
- **conjunto**: colección sin duplicados.

Esto equivale a decidir cómo se representa formalmente una unidad de observación o una colección de observaciones.


```python
# Lista: colección ordenada de observaciones simples
districts_cusco = ["Cusco", "Santiago", "San Sebastián", "San Jerónimo", "Wanchaq"]

print("Distritos:")
for district in districts_cusco:
    print("-", district)
```

```python
# Diccionario: una observación estructurada
household = {
    "hogar_id": "CUS_001",
    "distrito": "San Sebastián",
    "ingreso_mensual": 1850.0,
    "anios_educacion_jefe": 12,
    "acceso_agua_red": True
}

print("Registro del hogar:")
for key, value in household.items():
    print(f"{key}: {value}")
```

```python
# Tupla: útil cuando la estructura no debe modificarse
coordinate_cusco = (-13.53195, -71.96746)

print("Coordenadas de referencia:", coordinate_cusco)
print("Latitud :", coordinate_cusco[0])
print("Longitud:", coordinate_cusco[1])
```

```python
# Conjunto: útil para detectar categorías únicas
educational_levels = ["Primaria", "Secundaria", "Superior", "Secundaria", "Primaria", "Superior"]
unique_levels = set(educational_levels)

print("Niveles originales :", educational_levels)
print("Niveles únicos     :", unique_levels)
```

## 4. Lógica, comparación y control de flujo

Toda clasificación social hecha en código necesita reglas explícitas. Los condicionales permiten construir esas reglas. Esto es metodológicamente importante: una tipología computacional es una decisión conceptual formalizada.

```python
# Clasificación simple según ingreso mensual del hogar

household_income = 1850.0

if household_income < 1200:
    income_group = "bajo"
elif household_income < 2500:
    income_group = "medio"
else:
    income_group = "alto"

print("Ingreso del hogar:", household_income)
print("Grupo de ingreso :", income_group)
```

```python
# Uso de operadores lógicos

income = 1850.0
years_education = 12
has_water = True

is_relatively_stable = (income >= 1500) and (years_education >= 11) and has_water

print("¿Hogar relativamente estable?:", is_relatively_stable)
```

```python
# Bucle: aplicar una regla a múltiples observaciones

household_incomes = [950, 1400, 2100, 3200, 1100]

for i, value in enumerate(household_incomes, start=1):
    if value < 1200:
        label = "bajo"
    elif value < 2500:
        label = "medio"
    else:
        label = "alto"
    print(f"Hogar {i}: ingreso = {value} -> grupo = {label}")
```

## 5. Funciones: de la intuición a la formalización

Una función recibe entradas, aplica una regla y devuelve una salida. Es decir, hay un dominio de entrada; hay una regla de transformación; y hay un codominio de salida. En programación, una función bien escrita: hace una sola cosa; tiene entradas claras; devuelve una salida interpretable; y puede reutilizarse y verificarse.

```python
def classify_income(income: float) -> str:
    """
    Clasifica el ingreso mensual del hogar en tres grupos.

    Parámetros
    ----------
    income : float
        Ingreso mensual del hogar.

    Retorna
    -------
    str
        Etiqueta: 'bajo', 'medio' o 'alto'.
    """
    if income < 1200:
        return "bajo"
    elif income < 2500:
        return "medio"
    return "alto"


print(classify_income(900))
print(classify_income(1800))
print(classify_income(3200))
```

```python
def household_profile(income: float, education: int, water_access: bool) -> Dict[str, Any]:
    """
    Construye un perfil simplificado del hogar.
    """
    return {
        "income": income,
        "income_group": classify_income(income),
        "years_education": education,
        "water_access": water_access,
        "educational_threshold_met": education >= 11
    }


example_profile = household_profile(1850.0, 12, True)
example_profile
example_profile_1 = household_profile(800.0, 10, False)
example_profile_1
```

## 6. Primer paso hacia datos tabulares

Las encuestas y registros administrativos suelen representarse como tablas: filas y columnas. Una **fila** representa una unidad de observación; una **columna** representa una variable; y una **tabla** organiza sistemáticamente múltiples observaciones. Para nosotros, la tabla será una estructura central porque la Tarea 1 requerirá trabajar con módulos reales de datos.

```python
# Microdataset didáctico: hogares de Cusco
data = {
    "hogar_id": ["CUS_001", "CUS_002", "CUS_003", "CUS_004", "CUS_005"],
    "departamento": ["Cusco", "Cusco", "Cusco", "Cusco", "Cusco"],
    "anio": [2021, 2022, 2023, 2024, 2024],
    "ingreso_mensual": [980.0, 1450.0, 1750.0, 2100.0, 3100.0],
    "anios_educacion_jefe": [8, 11, 12, 14, 16],
    "participacion_ciudadana": [0, 1, 1, 0, 1],
    "acceso_agua_red": [True, True, True, False, True]
}

df = pd.DataFrame(data)
df
```

```python
# Inspección mínima rigurosa

print("Forma de la tabla:", df.shape)
print("\nTipos de datos:")
print(df.dtypes)

print("\nValores faltantes:")
print(df.isna().sum())

print("\nVista general:")
display(df.head())
```

```python
# Enriquecer la tabla con funciones definidas por nosotros

df["grupo_ingreso"] = df["ingreso_mensual"].apply(classify_income)
df["educacion_alta"] = df["anios_educacion_jefe"] >= 11

display(df)
```

```python
# Primeras preguntas analíticas

mean_income = df["ingreso_mensual"].mean()
share_high_education = df["educacion_alta"].mean()
share_participation = df["participacion_ciudadana"].mean()

print(f"Ingreso promedio del microdataset: {mean_income:.2f}")
print(f"Proporción con educación >= 11 años: {share_high_education:.2%}")
print(f"Proporción con participación ciudadana: {share_participation:.2%}")
```

```python
# Filtrado: ¿qué hogares combinan menor ingreso y menor escolaridad?

subset = df[
    (df["ingreso_mensual"] < 1800) &
    (df["anios_educacion_jefe"] < 11)
]

display(subset)
```
